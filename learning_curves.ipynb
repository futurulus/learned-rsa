```python
from matplotlib import pyplot as plot
```

```python
def load_series(filename, prefix):
    series = []
    with open(filename, 'r') as infile:
        for line in infile:
            if line.startswith(prefix):
                series.append(float(line[len(prefix):].strip()))
    return series
```

```python
en = load_series('error_curve_noada.csv', ' ')
ea = load_series('error_curve_ada.csv', '')
en0_1 = load_series('noada0.1.log', 'Error: ')
plot.plot(range(0, 50), en, range(0, 50), ea, range(len(en0_1)), en0_1)
plot.legend(['no-ada', 'adagrad', 'no-ada-0.1'])
plot.show()
```
